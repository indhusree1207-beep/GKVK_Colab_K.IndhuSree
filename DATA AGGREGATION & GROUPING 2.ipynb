```python
import pandas as pd

# 1. Create Sample Agricultural Data
data = {
    'Region': ['North', 'South', 'North', 'East', 'South', 'North'],
    'Crop': ['Wheat', 'Corn', 'Wheat', 'Rice', 'Corn', 'Wheat'],
    'Yield_Tons': [150, 200, 170, 300, 210, 160],
    'Acres': [50, 80, 55, 100, 85, 52]
}
df = pd.DataFrame(data)
print(df)
```

**.groupby()**

```python
df.groupby(['Region','Crop'])
```

```python
df.groupby(['Region','Crop']).sum()
```

```python
df.groupby(['Yield_Tons','Acres']).sum()
```

**AGGREGATION FUNCTIONS **

```python
df.groupby('Crop')['Yield_Tons'].sum()
```

```python
df.groupby('Region')['Crop'].sum()
```

```python
df.groupby('Crop')['Acres'].sum()
```

```python
df.groupby('Region').sum()
```

```python
df.groupby('Crop').sum()
```

```python
df.groupby('Yield_Tons').sum()
```

```python
df.groupby('Acres').sum()
```

**.mean()**

```python
df.groupby('Region')['Yield_Tons'].mean()
```

```python
df.groupby('Crop')['Yield_Tons'].mean()
```

```python
df.groupby('Region')['Yield_Tons'].count()
```

```python
#Combining enumerate() and zip()

crops  = ["Rice", "Wheat", "Maize"]
yields = [1200, 1500, 1100]

for plot_no, (crop, yield_kg) in enumerate(zip(crops, yields), start=1):
    print(f"Plot {plot_no}: {crop} → {yield_kg} kg")
```

```python
#Zip two dictionaries (DEFAULT behavior)
#By default, zip() iterates over dictionary keys

crop_price = {
    "Rice": 40,
    "Wheat": 35,
    "Maize": 30
}

crop_yield = {
    "Rice": 1200,
    "Wheat": 1500,
    "Maize": 1100
}

for a, b in zip(crop_price, crop_yield):
    print(a, b)
```

```python
#Zip dictionary KEYS with VALUES

for crop, price in zip(crop_price.keys(), crop_price.values()):
    print(crop, price)
```

```python
#Zip dictionary KEYS with VALUES

for crop, price, yields in zip(crop_price.keys(), crop_price.values(), crop_yield.values()): #here i also referred the crop yields to the values
    print(crop, price, yields)
```

```python
for crop, price, yields_kg in zip(
  crop_price.keys(),
  crop_price.values(),
  crop_yield.values()):
  print(f"{crop}:Price={price}, Yield={yields}")
```

```python
#Zip + Unzip with Dictionary Data

items = list(crop_price.items())
keys, values = zip(*items)

print(keys)
print(values)
```

```python
def collect_info():
  yields=[]
  n=int(input("Enter the number of fields:"))
  for i in range(n):
    value=int(input(f"Enter the crop yield of the field {i+1}(in quintals):"))
    yields.append(value)
  return yields

def average_crop_yield(yields):
  total_yield=sum(yields)
  return total_yield/len(yields)

def high_yield_field(yields,threshold):
  high_yield=[]
  for value in yields:
    if value>=threshold:
      high_yield.append(value)
  return sorted(high_yield)

def yield_statistics(yields):
  print("Total number of fields:",len(yields))
  print("Maximum crop yields:",max(yields))
  print("Minimum crop yield:",min(yields))
  print("Average crop yield:",average_crop_yield(yields))

def crop_management_system():
  print("======Crop Management System======")
  farmer_name=input("Enter the farmer name:")
  data=collect_info()
  print("Crop yield of different fields:",data)

  high=high_yield_field(data,500)
  print("Fields with high yield:",high)

  stats=yield_statistics(data)
  print(f"End of the program,Thank you {farmer_name}")

crop_management_system()
```

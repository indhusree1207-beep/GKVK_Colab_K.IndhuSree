```python
#AGRICULTURAL DATA ANALYSIS PROJECT

#------Function Definitions-----
def load_data(filename):
    """Reads crop data from a file and returns a list of dictionaries"""
    crops = []
    file =open(filename, "r")
    for line in file:
        line = line.strip()
        if line:
          data = line.split(",")
          crop = {
                 "name": data[0],
                 "season": data[1],
                 "area": float(data[2]),
                 "rainfall": float(data[3]),
                 "fertilizer":float(data[4]),
                 "yield": float(data[5])
                }
          crops.append(crop)
    return crops


def calculate_yield_per_acre(crop):
  """Returns yield per unit area"""
  return crop["yield"]/crop["area"]

def display_crops(crops):
  """Displays all crop records"""
  print("\nCROP RECORDS:")
  for crop in crops:
    ypa=calculate_yield_per_acre(crop)
    print(f"{crop['name']:10} | {crop['season']:8} | Yield/Area: {round(ypa,2)}")

def find_highest_and_lowest_yield(crops):
  """Finds crops with the highest and lowest yield"""
  if not crops:
    return None, None

  highest = crops[0]
  lowest = crops[0]

  for crop in crops:
    if calculate_yield_per_acre(crop) > calculate_yield_per_acre(highest):
      highest = crop
    if calculate_yield_per_acre(crop) < calculate_yield_per_acre(lowest):
      lowest = crop
  return highest,lowest

def seasonal_summary(crops):
  """Generates seasonal yield summary using dictionary"""
  summary={}
  for crop in crops:
    season=crop["season"]
    ypa=calculate_yield_per_acre(crop)
    if season not in summary:
      summary[season] = []
    summary[season].append(ypa)
  return summary

def write_report(filename, crops, highest, lowest, summary):
  """Writes analysis results to output file"""
  file=open(filename,"w")

  file.write("CROP YIELD PER UNIT AREA \n")

  for crop_item in crops:
    ypa_item = calculate_yield_per_acre(crop_item)
    file.write(f"{crop_item['name']}-{crop_item['season']}-Yield/Area:{round(ypa_item,2)}\n")

  file.write("\nHIGHEST YIELD\n")
  if highest:
    file.write(highest["name"]+"\n")
  else:
    file.write("N/A\n")

  file.write("\nLOWEST YIELD\n")
  if lowest:
    file.write(lowest["name"]+"\n")
  else:
    file.write("N/A\n")

  file.write("\nSEASONAL AVERAGE YIELD \n")
  for season in summary:
    if summary[season]:
      avg=sum(summary[season]) / len(summary[season])
      file.write(f"{season}: {round(avg, 2)}\n")
    else:
      file.write(f"{season}: No data\n")

  file.close() # Added to ensure the file is closed

#-----Main Program-----
def main():
  input_file="cropinfo.txt"
  output_file="analysis_report.txt"

  crops=load_data(input_file)

  if not crops:
    print("No crop data loaded for analysis.")
    return

  display_crops(crops)

  highest,lowest=find_highest_and_lowest_yield(crops)

  summary=seasonal_summary(crops)
  print("SEASONAL SUMMARY:", summary)

  write_report(output_file,crops,highest,lowest,summary)
  print("\nANALYSIS COMPLETE. REPORT WRITTEN TO",output_file)

main()
```

```python
# ==========================================
# AGRICULTURAL DATA ANALYSIS PROJECT
# ==========================================

def load_data(filename):
    """Reads crop data from a file and returns a list of crop dictionaries"""
    crops = []
    with open(filename, "r") as file:
            for line in file:
                line = line.strip()
                if line:
                    data = line.split(",")
                    crop = {
                        "name": data[0],
                        "season": data[1],
                        "area": float(data[2]),
                        "rainfall": float(data[3]),
                        "yield": float(data[4])
                    }
                    crops.append(crop)
    return crops


def calculate_yield_per_acre(crop):
    """Returns yield per unit area safely"""
    return crop["yield"] / crop["area"]


def display_crops(crops):
  """Displays all crop records"""
  print("\nCROP RECORDS:")
  for crop in crops:
    ypa=calculate_yield_per_acre(crop)
    ("\n")
    print(f"{crop['name']:10} | {crop['season']:8} | Yield/Area: {round(ypa,2)}")
    print(crops)

def find_highest_and_lowest_yield(crops):
    """Finds crops with the highest and lowest yield per acre"""
    if not crops:
        return None, None

    highest = lowest = crops[0]

    for crop in crops:
        if calculate_yield_per_acre(crop) > calculate_yield_per_acre(highest):
            highest = crop
        if calculate_yield_per_acre(crop) < calculate_yield_per_acre(lowest):
            lowest = crop

    return highest, lowest


def seasonal_summary(crops):
    """Creates seasonal yield summary"""
    summary = {}
    for crop in crops:
        season = crop["season"]
        ypa = calculate_yield_per_acre(crop)
        summary.setdefault(season, []).append(ypa)
    return summary


def write_report(filename, crops, highest, lowest, summary):
    """Writes analysis report to a file"""
    with open(filename, "w") as file:
        file.write("AGRICULTURAL YIELD ANALYSIS REPORT\n")
        file.write("=" * 35 + "\n\n")

        file.write("CROP YIELD PER UNIT AREA\n")
        for crop in crops:
            ypa = calculate_yield_per_acre(crop)
            file.write(f"{crop['name']} ({crop['season']}): {ypa:.2f}\n")

        file.write("\nHIGHEST YIELD CROP\n")
        file.write(f"{highest['name']}\n")

        file.write("\nLOWEST YIELD CROP\n")
        file.write(f"{lowest['name']}\n")

        file.write("\nSEASONAL AVERAGE YIELD\n")
        for season, values in summary.items():
            avg = sum(values) / len(values)
            file.write(f"{season}: {avg:.2f}\n")

        file.write(f"\nTOTAL CROPS ANALYZED: {len(crops)}\n")


# ==========================================
# MAIN PROGRAM
# ==========================================

def main():
    input_file = "cropsss.txt"
    output_file = "analysis_report.txt"

    crops = load_data(input_file)

    for crop in crops:
      print(crop)
    for crop in crops:
      for(key,value) in crop.items():
        print(f"{key}:{value}")
    display_crops(crops)
    print(crops)

    highest, lowest = find_highest_and_lowest_yield(crops)
    summary = seasonal_summary(crops)

    write_report(output_file, crops, highest, lowest, summary)

    print("\nANALYSIS COMPLETE ✔")
    print("Report saved as:", output_file)


main()
```

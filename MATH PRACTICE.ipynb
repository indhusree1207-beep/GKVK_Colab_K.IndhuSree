```python
# Mathematical concepts applied to farming

import math

# 1.Simple Equation:Yiled Prediction
def predict_yield(rainfall,fertilizer,base_yield=2000):
  """
  Simple linear model: Yield=base + (rain effect) + (fert effect)
  This is the foundation of regression!
  """
  rain_coefficient = 10 # kg per mm of rain
  fert_coefficient = 15 #kg per unit of fertilizer

  predicted = base_yield + (rainfall*rain_coefficient) + (fertilizer * fert_coefficient)
  return predicted

print("=== Yield Predicition (Linear Model) ===")
print(f"100mm rain, 50 units fert: {predict_yield(100,50)} kg/hectare")

# 2. Optimization: Finding best input levels
def profit_function(fertilizer_amount,price_per_kg=20,fert_cost_per_unit=50):
    """Find fertilizer amount that maximizes profit
    Yield increases with fertilizer but with diminishing returns
    """
    #Diminishing returns: sqrt function'
    yield_increase = 500 *math.sqrt(fertilizer_amount)

    revenue= yield_increase * price_per_kg
    cost = fertilizer_amount * fert_cost_per_unit
    profit = revenue - cost

    return profit

print("\n === OPTIMIZATION: BEST FERTILIZER AMOUNT ===")
for fert in [10,25,50,100,150]:
  profit = profit_function(fert)
  print(f"Fertilizer: {fert} units -> Profit: ₹{profit:0f}")
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# 1. User input for 10 values
rainfall_list=[]
fertilizer_list=[]

print("Enter 10 rainfall values (in mm):")
for i in range(10):
  val = float(input(f"Rainfall {i+1}: "))
  rainfall_list.append(val)

print("Enter 10 fertilizer values (in units):")
for i in range(10):
  val = float(input(f"Fertilizer {i+1}: "))
  fertilizer_list.append(val)

def predict_yield(rainfall,fertilizer,base_yield=2000):
  rain_coefficient= 10 # kg per mm of rain
  fert_coefficient= 15 # kg per unit of fertilizer
  predicted=base_yield + (rainfall * rain_coefficient) + (fertilizer * fert_coefficient)
  return predicted

yield_list = [predict_yield(r,f) for r,f in zip(rainfall_list,fertilizer_list)]

def profit_function(fertilizer_amount,price_per_kg=20, fert_cost_per_unit=50):
    yield_increase = 500 * math.sqrt(fertilizer_amount)
    revenue = yield_increase * price_per_kg
    cost = fertilizer_amount * fert_cost_per_unit
    profit = revenue - cost

    return profit

profit_list = [profit_function(f) for f in fertilizer_list]

# Display results
print("\n ===RESULTS ===")
for i in range(10):
  print(f"Rainfall: {rainfall_list[i]} mm, Fertilizer: {fertilizer_list[i]} units -> Yield: {yield_list[i]:.2f} kg/ha, Profit: ₹{profit_list[i]:.2f}")
```

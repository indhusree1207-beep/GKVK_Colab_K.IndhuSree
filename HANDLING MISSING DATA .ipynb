FILLING MISSING VALUES

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 31, 22, None, 27],
    'Gender': ['F', 'M', None, 'M', 'F'],
    'Salary': [50000, np.nan, 30000, 40000, 60000]
}
df = pd.DataFrame(data)

df.fillna(0)  # Fills all the missing values with the value mentioned in the bracket

```

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 31, None, None, 27],
    'Gender': ['F', np.nan, None, 'M', 'F'],
    'Salary': [50000, np.nan, 30000, None, 60000]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df.fillna({'Age':"Unknown",'Gender':"Not mentioned","Salary":"Not specified"})
```

REPLACING VALUES

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 31, 22, None, 27],
    'Gender': ['F', 'M', None, 'M', 'F'],
    'Salary': [50000, np.nan, 30000, 40000, 60000]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

print("\n === NEW DATA FRAME ==  ")
df.replace('Eva','Emily')

```

```python
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'name':['Alice','Bob','Charlie','David','Eva','Frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})

sal=df.sort_values('salary')    # Ascending order(by default)
print(sal)
```

```python
sal=df.sort_values('experience')
print(sal)
```

```python
sal=df.sort_values('salary',ascending=False)   # Descending order
print(sal)
```

```python
sal=df.sort_values(['salary','experience',])
print(sal)
```

```python
sal=df.sort_values(['salary','experience'],ascending=[False,True])
print(sal)
```

```python
des=df.sort_index(ascending=False)
print(des)
```

```python
df['rank_salary']=df['salary'].rank()
print(df)
```

```python
df['rank_salary']=df['salary'].rank(ascending=False)
print(df)
```

```python
df['dense_rank_salary']=df['salary'].rank(ascending=False,method='dense')
print(df)
```

```python
df=df['department'].value_counts(normalize=True)
print(df)
```

PLOTTING GRAPHS

```python
import matplotlib.pyplot as plt
import numpy as np
df=pd.DataFrame({
    'name':['Alice','Bob','Charlie','David','Eva','Frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})

df[["salary","experience"]].plot()
plt.show()
```

```python
df[["experience","salary"]].plot()
plt.show()
```

```python
df.plot(kind='bar',x='name',y='salary')
plt.show()
```

```python
df['salary'].plot(kind='hist',bins=5)
plt.show()
```

```python
df['salary'].plot(kind='hist',bins=10)
plt.show()
```

```python
df.plot(kind='scatter',x='experience',y='salary')
plt.show()
```

```python
df[['salary','experience']].plot(kind='box')
plt.show()
```

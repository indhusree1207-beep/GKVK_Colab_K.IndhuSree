```python
import matplotlib as plt
import numpy as np


rainfall=[]
yields=[]
print("===== RAINFALL V/S CROP YIELDS =====")
number_of_months=int(input("Enter the number of months:"))
for i in range(number_of_months):
  rainfalls=float(input(f"Enter the rainfall (in mm) of month {i+1}:"))
  yields_m=float(input(f"Enter the yield (in tons/ha) of month {i+1}:"))

  rainfall.append(rainfalls)
  yields.append(yields_m)

total_rainfall=np.array(rainfall)
yields=np.array(yields)

print("\n ")
print("===== RAINFALL STATISTICS =====")
print(f"Mean:{np.mean(total_rainfall):.2f} mm")
print(f"Median:{np.median(total_rainfall):.2f} mm")
unique_rainfall_values, counts = np.unique(total_rainfall, return_counts=True)
mode_rainfall = unique_rainfall_values[np.argmax(counts)]
print(f"Mode:{mode_rainfall:.2f} mm")
print(f"Std Dev:{np.std(total_rainfall):.2f} mm")
print(f"Variance:{np.var(total_rainfall):.2f} mm")
print(f"CoVarinace:{np.cov(total_rainfall,yields)[0,1]:.2f}")
print(f"Minimum:{np.min(total_rainfall)} mm (Month{np.argmin(total_rainfall)+1})")
print(f"Maximum:{np.max(total_rainfall)} mm (Month{np.argmax(total_rainfall)+1})")
print(f"Range:{np.ptp(total_rainfall)} mm")

print("\n ===== YIELD STATISTICS =====")
print(f"Mean:{np.mean(yields):.2f} tons/ha")
print(f"Median:{np.median(yields):.2f} tons/ha")
print(f"Std Dev:{np.std(yields):.2f} tons/ha")

print("\n ===== PERCENTILES =====")
percentiles=[25,50,75,90]
rainfall_percentiles=np.percentile(total_rainfall,percentiles)
for p,val in zip(percentiles,rainfall_percentiles):
  print(f"{p}th percentile rainfall:{val:.2f} mm")

print("\n ===== CORRELATION ANALYSIS ===== ")
correlation_matix=np.corrcoef(total_rainfall,yields)
correlation=correlation_matix[0,1]
print(f"Rainfall-Yield Correlation:{correlation:.4f}")
print("Interpretation:Strong postitive correlation" if correlation >0.7 else "Moderate Correlation" if correlation >0.4 else "Weak Correlation" )

print("\n ===== CUMULATIVE RAINFALL ===== ")
cumulative=np.cumsum(total_rainfall)
print("Month\tRainfall\tCumualative")
for month,(rain,cum) in enumerate(zip(total_rainfall,cumulative),1):
  print(f"{month}\t{rain}\t\t{cum}")25
```

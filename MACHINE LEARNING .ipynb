```python
# STep 1 : Import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , r2_score

# Step 2 : Create Sample Agricultural Dataset
data={
        "Rainfall": [450, 520, 390, 610, 480, 530, 410, 590],
    "Temperature": [22, 20, 25, 19, 23, 21, 24, 18],
    "Nitrogen": [0.8, 1.0, 0.6, 1.2, 0.9, 1.1, 0.7, 1.3],
    "Fertilizer": [120, 140, 100, 160, 130, 150, 110, 170],
    "Yield": [3.5, 4.2, 2.9, 4.8, 3.8, 4.4, 3.1, 5.0]
}

df=pd.DataFrame(data)

# Step 3 : Define Features (X) and Target (y)
X=df[["Rainfall","Temperature","Nitrogen","Fertilizer"]]
y=df["Yield"]

print(X)
print(y)

# Step 4 : Split DAta into Training and Testing Sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)

print(X_train)
print(X_test)

# Step 5: Train Model
model=LinearRegression()
model.fit(X_train,y_train)


# Step 6  : Make Predictions
y_pred=model.predict(X_test)

# Step 7 : Evaluate Model
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

print("Mean Absolute Error :",round(mae,2))
print("R2 Score:",round(r2,2))

#  Step 8 : Predict Yield for New Farm Conditions
new_data =pd.DataFrame({"Rainfall":[500],"Temperature":[21],"Nitrogen":[1.0],"Fertilizer":145})
predicted_yield=model.predict(new_data)

print("Predicted Yield(tons/ha):",round(predicted_yield[0],2))


# Step 1: Import libraries: It imports necessary libraries like numpy for numerical operations, pandas for data manipulation, train_test_split for splitting data, LinearRegression for the model, and mean_absolute_error and r2_score for model evaluation.
# Step 2: Create Sample Agricultural Dataset: A sample dataset is created as a dictionary, containing features like 'Rainfall', 'Temperature', 'Nitrogen', 'Fertilizer', and the target variable 'Yield'. This dictionary is then converted into a pandas DataFrame.
# Step 3: Define Features (X) and Target (y): The DataFrame is separated into input features (X), which are 'Rainfall', 'Temperature', 'Nitrogen', and 'Fertilizer', and the target variable (y), which is 'Yield'.
# Step 4: Split Data into Training and Testing Sets: The data is divided into training and testing sets using train_test_split. This is crucial to evaluate how well the model generalizes to unseen data. 75% of the data is used for training and 25% for testing.
# Step 5: Train Model: A LinearRegression model is initialized and then trained (fit) using the training features (X_train) and training target (y_train).
# Step 6: Make Predictions: The trained model (model) is used to make predictions (predict) on the unseen test features (X_test).
# Step 7: Evaluate Model: The model's performance is assessed using two metrics: Mean Absolute Error (MAE), which measures the average absolute difference between actual and predicted values, and R2 Score, which indicates the proportion of the variance in the dependent variable that is predictable from the independent variables.
# Step 8: Predict Yield for New Farm Conditions: Finally, the code demonstrates how to use the trained model to predict the yield for new, hypothetical farm conditions (e.g., specific rainfall, temperature, nitrogen, and fertilizer values).
```

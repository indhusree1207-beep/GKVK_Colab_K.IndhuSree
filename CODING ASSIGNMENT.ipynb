```python
# TASK 2
rainfall=[]
number_of_crops=int(input("Enter the number of crops:"))
for i in range(number_of_crops):
  rf=float(input(f"\n Enter the amount on rainfall of crop {i+1} (in mm) :"))
  rainfall.append(rf)

highest=max(rainfall)
lowest=min(rainfall)
average=sum(rainfall)/len(rainfall)

print("Highest Rainfall:",highest,"mm")
print("Lowest Rainfall:",lowest,"mm")
print("Average Rainfall:",round(average,2),"mm")
```

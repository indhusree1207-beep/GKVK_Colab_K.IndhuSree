```python
my_tuple = (1, 2, 3, 'apple', 'banana')
print("Original tuple:", my_tuple)
print("Type of my_tuple:", type(my_tuple))

# Tuples are immutable: you cannot change elements (this will cause an error)
# Uncomment the line below to see the TypeError
# my_tuple[0] = 100

# You can access elements by index
print("First element of tuple:", my_tuple[0])

# Slicing a tuple
print("Slicing tuple (elements from index 1 to 3):", my_tuple[1:4])

# You can create a new tuple by concatenating
new_tuple = my_tuple + ('cherry',)
print("New tuple :", new_tuple)
```

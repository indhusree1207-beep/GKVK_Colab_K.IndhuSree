```python
# Create a dictionary
my_dict = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

print("Original dictionary:", my_dict)

# Accessing values using keys
print("Name:", my_dict['name'])
print("Age:", my_dict['age'])

# Modifying a value
my_dict['age'] = 31
print("Dictionary after modifying age:", my_dict)

# Adding a new key-value pair
my_dict['occupation'] = 'Engineer'
print("Dictionary after adding occupation:", my_dict)

# Deleting a key-value pair
del my_dict['city']
print("Dictionary after deleting city:", my_dict)

# Iterating through a dictionary
print("\nIterating through keys:")
for key in my_dict:
    print(key)

print("\nIterating through values:")
for value in my_dict.values():
    print(value)

print("\nIterating through key-value pairs:")
for key, value in my_dict.items():
    print(f"{key}: {value}")
```

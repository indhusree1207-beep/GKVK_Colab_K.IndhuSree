```python
import matplotlib as plt
import numpy as np

import numpy as np
rainfall=np.array([45,52,68,120,180,220,195,160,110,75,58,40])
yields=np.array([2.1,2.5,2.8,4.2,5.8,6.5,5.9,5.2,3.8,2.9,2.4,2.0])

print("\n ===== OTHER STATISTICAL OPERATIONS =====")
min_val_r=np.min(rainfall)
min_val_y=np.min(yields)
max_val_r=np.max(rainfall)
max_val_y=np.max(yields)
argmin_r=np.argmin(rainfall)
argmax_r=np.argmax(rainfall)
argmin_y=np.argmin(yields)
argmax_y=np.argmax(yields)
print("\n MINIMUM RAINFALL:",min_val_r)
print("\n MAXIMUM RAINFALL:",max_val_r)
print("\n MINIMUM YIELD:",min_val_y)
print("\n MAXIMUM YIELD:",max_val_y)
print("\n INDEX OF MIN R.F:",argmin_r)
print("\n INDEX OF MAX R.F:",argmax_r)
print("\n INDEX OF MIN YIELD:",argmin_y)
print("\n INDEX OF MAX YIELD:",argmax_y)

r25=np.percentile(rainfall,25)
y25=np.percentile(yields,25)
print("\n 25th PERCENTILE OF R.F",r25)
print("\n 25th PERCENTILE OF YIELD",y25)

r50=np.percentile(rainfall,50)
y50=np.percentile(yields,50)
print("\n 50th PERCENTILE OF R.F",r50)
print("\n 50th PERCENTILE OF YIELD",y50)

r75=np.percentile(rainfall,75)
y75=np.percentile(yields,75)
print("\n 75th PERCENTILE OF R.F",r75)
print("\n 75th PERCENTILE OF YIELD",y75)

r100=np.percentile(rainfall,100)
y100=np.percentile(yields,100)
print("\n 100th PERCENTILE OF R.F",r100)
print("\n 100th PERCENTILE OF YIELD",y100)

print("\n ===== WEIGHTED AVERAGE =====")
random = [88, 92, 85, 90]
numbers= [1,2,3,4]
weighted_average = np.average(random, weights=numbers)
print("\n AVERAGE",weighted_average)
```

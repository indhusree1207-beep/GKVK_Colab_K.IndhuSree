In pandas, a Series is a one-dimensional, labeled array used to hold a single column of data,
while a DataFrame is a two-dimensional, tabular data structure with rows and columns

```python
import pandas as pd
s1=pd.Series([10,20,30],index=["a","b","c"])
s2=pd.Series([40,50,60],index=["a","b","c"])
df=pd.DataFrame({"Col 1":s1,"Col 2":s2})
print(df)
```

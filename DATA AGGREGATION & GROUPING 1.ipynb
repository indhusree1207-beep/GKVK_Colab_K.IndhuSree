```python
import pandas as pd

data={
    'Department':['HR','Sales','IT','Sales','IT','HR','Sales'],
    'Employee':['A','B','C','D','E','F','G'],
    'Salary':[50000,60000,45000,48000,70000,72000,71000],
    'Experience':[2,4,3,5,6,7,5]
}

df=pd.DataFrame(data)
print(df)
```

**.groupby()**

```python
df1=df.groupby(['Department','Experience'])
print(df1)
```

**.sum()**

```python
df1=df.groupby(['Department','Experience']).sum()
print(df1)
```

```python
df2=df.groupby('Experience').sum()
print(df2)
```

**AGGREGATION FUNCTIONS **

```python
df.groupby('Department')['Salary'].sum()
```

```python
df.groupby('Department').sum()
```

```python
df.groupby('Salary')['Department'].sum()
```

**.mean()**

```python
df.groupby('Department')['Salary'].mean()
```

**.count()**

```python
df.groupby('Department')['Employee'].count()
```

**MULTIPLE AGGREGATIONS **

```python
df.groupby("Department")["Salary"].agg(['sum','mean','std'])
```

**CUSTOM AGGREGATIONS **

```python
df.groupby('Department').agg({
    "Salary":'sum', 'Experience':'mean'
})
```

**BROADCAST GROUP MEAN TO EACH ROW**

```python
df['Dept_Avg_Salary'] = df.groupby('Department')['Salary'].transform('mean')
print(df['Dept_Avg_Salary'])
```

```python
df['Dept_Avg_Salary'] = df.groupby('Department')['Salary'].transform('mean')
df4=df['Dept_Avg_Salary']
print(df4)
```

**FILTERING GROUPS **

```python
df.groupby('Department').filter(lambda x : len(x)>2)
```

```python
df5=df.groupby('Department').filter(lambda x : len(x)>2)
print(df5)
```

**PIVOT TABLES **

```python
#Mean salary by dept and experience

df.pivot_table(
    values='Salary',
    index='Department',
    columns='Experience',
    aggfunc='mean'
)
```

```python
# IMPORT LIBRARIES
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB

# CREATE DATASET
data = {
'Weather': ['Sunny', 'Sunny', 'Rainy', 'Rainy', 'Cloudy', 'Sunny', 'Cloudy', 'Rainy'],
'Soil': ['Dry', 'Wet', 'Wet', 'Dry', 'Dry', 'Dry', 'Wet', 'Wet'],
'Temperature': ['Hot', 'Hot', 'Cool', 'Cool', 'Cool', 'Cool', 'Hot', 'Hot'],
'Irrigate': ['Yes', 'No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'No']
}
df = pd.DataFrame(data)
print("Dataset:In", df)

# ENCODE CATEGORICAL FEATURES
le_weather=LabelEncoder()
le_soil=LabelEncoder()
le_temp=LabelEncoder()
le_target=LabelEncoder()
df['Weather'] = le_weather.fit_transform(df['Weather'])
df['Soil'] = le_soil.fit_transform(df['Soil'])
df['Temperature'] = le_temp.fit_transform(df['Temperature'])
df['Irrigate'] = le_target.fit_transform(df['Irrigate'])
print("InEncoded Dataset:In", df)

# SPLIT FEATURES AND TARGET
X=df[["Weather","Soil","Temperature"]]
y=df["Irrigate"]

# TRAIN NAIVE BAYES MODEL
model=CategoricalNB()
model.fit(X,y)

# TEST INPUT (SUNNY,DRY,COOL)
test_data=pd.DataFrame({
    'Weather': ['Sunny'],
'Soil': ['Dry'],
'Temperature': ['Cool']
})

# ENCODE TEST DATA USING SAME ENCODERS
test_data['Weather'] = le_weather.transform(test_data['Weather'])
test_data['Soil']= le_soil.transform(test_data['Soil'])
test_data['Temperature'] = le_temp.transform(test_data['Temperature'])

# PREDICTION
prediction=model.predict(test_data)

# CONVERT BACK TO ORIGINAL LABEL
result=le_target.inverse_transform(prediction)

print("\n TEST INPUT : SUUNNY , DRY , COOL")
print("PREDICTION:",result[0])

# PROBABILITY OUTPUT (optional)
prob=model.predict_proba(test_data)
print("CLASS PROBABLITIES:",prob)
```

```python
another_nested_list = [
    [10, 20, 30],
    ['apple', 'banana', 'cherry'],
    [True, False, None, {"key": "value"}],
    [[1, 2], [3, 4]]
]

print(another_nested_list)
print(another_nested_list[1])
print(another_nested_list[2][3])
print(another_nested_list[3][0][1])
```

```python
nested_list = [
    [1, 2, 3],
    ['a', 'b', 'c'],
    [True, False]
]

print(nested_list)
print(nested_list[0])
print(nested_list[1][1])
```

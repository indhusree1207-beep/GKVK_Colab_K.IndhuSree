```python
import numpy as np
x=np.arange(10)
x=x[::2]*2    # x[::2] selects every second element of the array x, starting from the first element (index 0). This results in a temporary array [0, 2, 4, 6, 8].
              # *2 then multiplies each element of this temporary array by 2, yielding [0, 4, 8, 12, 16].
print(x)
```

Suppose a has shape (2,3) and b has shape (3,). What is the shape of a + b under NumPy broadcasting rules, assuming the operation is valid?

ANSWER:
*NumPy broadcasting starts by comparing shapes from the rightmost (trailing) dimension. If the arrays have a different number of dimensions, the shape of the smaller array is padded with 1s on its left side.
Shape of a=(2,3)
Shape of b=(3,) prepended with 1 becomes (1,3)
*Two dimensions are compatible if they are equal, or if one of them is 1.
*Trailing dimension: Both are 3.They match.
Leading dimension a has 2 and b(padded) has 1
. Since one is 1 ,they are compatible.
*Dimension 1: max(2,1)=2
Dimension 2:max(3,3)=3



```python
import numpy as np
a = np.array([[1, 2], [3, 4], [5, 6]])
a[a[:, 1] % 2 == 0]
# a[:, 1]: This selects all elements from the second column of the array a, resulting in [2, 4, 6].
# % 2 == 0: This performs a modulo operation on the selected column and checks if each element is evenly divisible by 2, generating a boolean array: [True, True, True].
# a[...]: This uses the resulting boolean array to select only the rows in a where the corresponding boolean value is True.
```

```python
import numpy as np
a = np.array([[1, 2], [3, 4], [5, 6]])
a[:, a[1, :] % 2 == 0]
# The code attempts to use boolean array indexing where the indices applied to the first axis of array a are generated from the second axis. This causes an IndexError because the shape of the boolean mask does not match the axis it is being applied to [1].
# Specifically: a has the shape (3, 2).
# a[1, :] % 2 == 0 evaluates to a 1-dimensional boolean array of shape (2,), which is [False, True].
# The slicing a[:, ...] attempts to apply this 2-element mask to the first axis of a, which has 3 elements.
```

```python
import numpy as np
a = np.array([[1, 2], [3, 4], [5, 6]])
a[a[:, 0] % 2 == 0, 1]
# The code a[a[:, 0] % 2 == 0, 1] selects elements from the second column (index 1) of the NumPy array a where the corresponding element in the first column (index 0) is even.
# a[:, 0]: This slices the array to get all elements in the first column: [1, 3, 5].
# a[:, 0] % 2 == 0: This applies a modulo operation and a comparison to create a boolean mask, which evaluates to [False, False, False] because all elements in the first column are odd.
# a[..., 1]: This attempts to use the boolean mask to select rows and then retrieve the second column.
```

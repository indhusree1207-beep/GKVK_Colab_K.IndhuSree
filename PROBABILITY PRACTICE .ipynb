```python
# Statistics with Farm Data
import statistics

#Annual yields(kg/ha) for past years
yields=[2500,2800,2200,3000,2700]

print("===== YIELD STATISTICS ===telek=")
print(f"Yields: {yields}")
print(f"Mean:{statistics.mean(yields):.0f}kg/ha")
print(f"Std Deviation:{statistics.stdev(yields):.0f}kg/ha")
print(f"Range:{max(yields) - min(yields):.0f}kg/ha")

# Probability calculation
total_years=10
drought_years=3
prob_drought=drought_years/total_years
print(f"\n DROUGHT PROBABILTY:{prob_drought:.0%}")

# Variance Interpretation
if statistics.stdev(yields)>300:
  print("⚠️HIGH VARIABILITY - Consider Crop Insurance")
else:
    print("✅ STABLE YIELDS - Good Predictability")
```

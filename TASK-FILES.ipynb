```python
def crop_data(filename):
   crops=[]
   file=open(filename,"r")
   for line in file:
       line=line.strip()
       if line:
          data=line.split(",")
          crop={
               "name":data[0],
               "season":data[1],
               "area":float(data[2]),
               "yield":float(data[3]),
               "cost":float(data[4]),
               "revenue":float(data[5])
             }
          crops.append(crop)
   file.close()
   return crops

crop_data("crop_profit.txt")
```

```python
def find_all_kharif_and_non_kharif_crops(crops):
  seasonal_summary = {}
  for crop in crops:
      season = crop["season"]
      if season not in seasonal_summary:
          seasonal_summary[season] = []
      seasonal_summary[season].append(crop["name"])
  return seasonal_summary

seasonal_summary = find_all_kharif_and_non_kharif_crops(crops)
print(seasonal_summary)
```

```python
def find_all_kharif_and_non_kharif_crops(crops):
  seasonal_summary={}
  kharif=[]
  non_kharif=[]
  for crop in crops:
    if crop["season"]=="Kharif":
      kharif.append(crop["name"])
    else:
      non_kharif.append(crop["name"])
  return kharif,non_kharif

crops = crop_data("crop_profit.txt")
seasonal_summary = find_all_kharif_and_non_kharif_crops(crops)
print(seasonal_summary)
print("KHARIF CROPS:",seasonal_summary[0])
print("NON-KHARIF CROPS:",seasonal_summary[1])
```

```python
def calculate_profit(crops):
  for crop in crops:
    profit = crop["revenue"] - crop["cost"]
    print(crop["name"],"|",crop["season"],"|","Profit:",profit)
calculate_profit(crops)
```

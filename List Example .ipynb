```python
my_list = [1, 2, 3, 'apple', 'banana']
print("Original list:", my_list)
print("Type of my_list:", type(my_list))

# Lists are mutable: you can change elements
my_list[0] = 100
print("List after changing an element:", my_list)

# You can also add elements
my_list.append('cherry')
print("List after appending an element:", my_list)

# Slicing a list
print("Slicing list (elements from index 1 to 3):", my_list[1:4])
```

```python
import numpy as np
# Crop yields for 4 farms over 5 years (tons/hectare)
yeilds=np.array([
    [45, 48, 52, 49, 53],
    [38, 41, 39, 44, 42],
    [52, 55, 58, 56, 60],
    [41, 43, 45, 47, 46]
])
# Weather impact factors for each year (1D array - broadcasts across farms)
weather_factors=np.array([1.0,0.95,1.05,0.98,1.02])

# Broadcasting: adjust all farm yields by weather factors
adjusted_yields=yeilds*weather_factors
print("\n WEATHER-ADJUSTED YIELDS:")
print(adjusted_yields.round(2))

# Boolean Indexing: find exceptional years
exceptional=adjusted_yields>55
print("\n EXCEPTIONAL YIELDS(>55):")
print(adjusted_yields[exceptional])

# Conditionsal Operations: Classify yields
categories=np.where(
    adjusted_yields>50 , 'High',
    np.where(adjusted_yields>40,'Medium','Low')
    )
print("\n YIELD CATEGORIES:")
print(categories)

# Advanced indexing: Select specific farms
selected_farms=[0,3] # Farm 1 and Farm 3
print(f"\n SELECTED FARMS YIELDS:")
print(yeilds[selected_farms])

# Multi-condition filtering
good_years=adjusted_yields[(adjusted_yields>45) & (adjusted_yields<55)]
print(f"\n GOOD PERFROMING YEARS(45-55 tons):")
print(good_years)
```

```python
import numpy as np
# Create a 10X5 array with values 0 to 49
a=np.arange(50).reshape(10,5)
print("\n ORIGINAL ARRAY(10X5):\n ",a)
```

```python
import numpy as np

a=np.arange(21).reshape(3,7)
print("\n ANOTHER ARRAY:\n",a)

print("\n TRIAL 1 \n ",(a[:]))  # This will print all the elements of the matrix in the same order
print("\n TRIAL 2 \n ",a[:,0])  # This will print the eleents that are present in the 0th index
print("\n TRIAL 3 \n ", a[:,1]) # In the same manner this will print the elemnts present in the 1st index
print("\n TRIAL 4 \n ",a[a[:,0]>10])   # This performs boolean indexing. It first creates a boolean array by checking which elements in the first column (a[:,0]) are greater than 10.
                                      # Then, it uses this boolean array to select and print only those rows from a where the condition was True.
```

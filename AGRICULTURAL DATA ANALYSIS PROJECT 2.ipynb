```python
#AGRICULTURAL DATA ANALYSIS PROJECT

#------Function Definitions-----
def load_data(filename):
    """Reads crop data from a file and returns a list of dictionaries"""
    crops = []
    file =open(filename, "r")
    for line in file:
        line = line.strip()
        if line:
          data = line.split(",")
          crop = {
             "name": data[0],
             "season": data[1],
             "area": float(data[2]),
             "rainfall": float(data[3]),
             "fertilizer": float(data[4]),
             "yield": float(data[5])
            }
        crops.append(crop)
    file.close()
    return crops


def calculate_yield_per_acre(crop):
  """Returns yield per unit area"""
  return crop["yield"]/crop["area"]

def display_crops(crops):
  """Displays all crop records"""
  print("\nCROP RECORDS:")
  for crop in crops:
    ypa=calculate_yield_per_acre(crop)
    print("\n")
    print(f"{crop['name']:10} | {crop['season']:8} | Yield/Area: {round(ypa,2)}")
    print(crop)

def find_highest_and_lowest_yield(crops):
  """Finds crops with the highest and lowest yield"""
  highest = crops[0]
  lowest = crops[0]

  for crop in crops:
    if calculate_yield_per_acre(crop) > calculate_yield_per_acre(highest):
      highest = crop
    if calculate_yield_per_acre(crop) < calculate_yield_per_acre(lowest):
      lowest = crop
  return highest,lowest

def best_crop_per_season(crops):
    best = {}
    for crop in crops:
        season = crop["season"]
        if season not in best:
            best[season] = crop
        else:
            if calculate_yield_per_acre(crop) > calculate_yield_per_acre(best[season]):
                best[season] = crop
    return best


def fertilizer_efficiency(crop):
  return crop["fertilizer"]/crop["area"]

def most_fertilizer_efficient_crop(crops):
    best= crops[0]
    for crop in crops:
      if fertilizer_efficiency(crop) > fertilizer_efficiency(best):
        best = crop
    return best

def rainfall_efficiency_anaysis(crops,max_rainfall):
  print("Crops with rainfall<=",max_rainfall)
  for crop in crops:
    if crop["rainfall"] <= max_rainfall:
      print(crop["name"],"-",crop["yield"])

def find_max_rainfall(crops):
  max_rainfall=crops[0]["rainfall"]
  for crop in crops:
    if crop["rainfall"]>max_rainfall:
      max_rainfall=crop["rainfall"]
  return max_rainfall

def crops_in_multiple_seasons(crops):
  crop_seasons={}
  for crop in crops:
    name=crop["name"]
    season=crop["season"]
    if name not in crop_seasons:
       crop_seasons[name]=[]
    if season not in crop_seasons[name]:
       crop_seasons[name].append(season)
  for name,seasons in crop_seasons.items():
    if len(seasons)>1:
      print(name,"grown in seasons:",seasons)

def seasonal_summary(crops):
  """Generates seasonal yield summary using dictionary"""
  summary={}
  for crop in crops:
    season=crop["season"]
    ypa=calculate_yield_per_acre(crop)
    if season not in summary:
      summary[season] = []
    summary[season].append(ypa)
  return summary

def rank_crops_by_yield(crops):
  ranked=sorted(crops,key=calculate_yield_per_acre,reverse=True)
  return ranked


def crop_count_per_season(crops):
  count={}

  for crop in crops:
    season=crop["season"]
    if season not in count:
      count[season]=0
    count[season]+=1
  return count

def write_report(filename, crops, highest, lowest, summary, count, most_fertilizer_efficient_crop_result, max_rainfall, ranked_crops_list, best_crops_per_season,crops_in_mulltiple_seasons):
  """Writes analysis results to output file"""
  file=open(filename,"w")

  file.write("CROP YIELD PER UNIT AREA \n\n")


  for crop_item in crops:
    ypa_item = calculate_yield_per_acre(crop_item)
    file.write(f"{crop_item['name']}-{crop_item['season']}-Yield/Area:{round(ypa_item,2)}\n")

  file.write("\nHIGHEST YIELD\n")
  file.write(highest["name"]+"\n")

  file.write("\nLOWEST YIELD\n")
  file.write(lowest["name"]+"\n")

  file.write("\n CROP COUNT PER SEASON\n")
  for season,value in count.items():
    file.write(f"{season}:{value}\n")

  file.write("\nMOST FERTILIZER EFFICIENT CROP\n")
  file.write(most_fertilizer_efficient_crop_result["name"] + "\n")

  file.write("\n RANKING CROPS BY PRODUCTIVITY\n")
  rank=1
  for crop in ranked_crops_list:
    file.write(f"{rank}.{(crop['name'])} -{round(calculate_yield_per_acre(crop),2)}\n")
    rank+=1

  file.write("\n CROPS GROWN IN MULTIPLE SEASONS \n")
  for crop in crops:
     file.write(name,"grown in seasons:",seasons)



  file.write("\n BEST PERFROMING CROP PER SEASON: \n")
  for season,crop_data in best_crops_per_season_dict.items():
    file.write(f"{season} : {crop_data["name"]}\n")


  file.write("\nCROPS WITH RAINFALL<="+str(max_rainfall)+"\n")

  file.write("\nSEASONAL AVERAGE YIELD \n")
  for season in summary:
    avg=sum(summary[season]) / len(summary[season])
    file.write(f"{season}: {round(avg, 2)}\n")


  file.close()
#-----Main Program-----
def main():
  input_file="cropdata.txt"
  output_file="analysis_report.txt"

  crops=load_data(input_file)

  for crop in crops:
    print(crop)
  for crop in crops:
    for(key,value) in crop.items():
      print(f"{key}:{value}")
  display_crops(crops)
  print(crops)

  count=crop_count_per_season(crops)
  print(count)

  highest,lowest=find_highest_and_lowest_yield(crops)

  most_fertilizer_efficient_crop_result=most_fertilizer_efficient_crop(crops)
  print("\n MOST FERTILIZER EFFICIENT CROP:\n",most_fertilizer_efficient_crop_result["name"])

  max_rain=find_max_rainfall(crops)

  ranked_crops=rank_crops_by_yield(crops)

  rainfall_efficiency_anaysis(crops, max_rain)

  best_crops_per_season=best_crop_per_season(crops)

  crop_seasons=crops_in_multiple_seasons(crops)

  summary=seasonal_summary(crops)

  print("SEASONAL SUMMARY (for reporting):", summary)

  write_report(output_file,crops,highest,lowest,summary, count, most_fertilizer_efficient_crop_result,max_rain,ranked_crops,best_crops_per_season,crops_in_multiple_seasons)
  print("\nANALYSIS COMPLETE. REPORT WRITTEN TO",output_file)

main()
```

```python
def add(x,y):
  return x+y
def subtract(x,y):
  return x-y
def multiply(x,y):
  return x*y
def divide (x,y):
  if y==0:
    print("Error!Division by zero")
  else:
    return x/y

def get_input():
  try:
    x=float(input("Enter the first number:"))
    y=float(input("Enter the second number:"))
    return x,y
  except ValueError:
    print("Invalid input!Please enter valid numbers.")
    return get_input()

def print_menu():
  print("Select operation:")
  print("1.Add")
  print("2.Subtract")
  print("3.Multiply")
  print("4.Divide")
  print("5.Exit")

def calculator():
  while True:
    print_menu()
    choice=input("Enter choice(1/2/3/4/5):")

    if choice == "5":
      print("Exiting the calculator.Goodbye!")
      break
    elif choice in ('1', '2', '3', '4'): # This handles valid operation choices
      x, y = get_input()
      if choice=='1':
        print(f"{x}+{y}={add(x,y)}")
      elif choice=='2':
        print(f"{x}-{y}={subtract(x,y)}")
      elif choice=='3':
        print(f"{x}*{y}={multiply(x,y)}")
      elif choice=='4':
        result = divide(x,y)
        if result is not None:
            print(f"{x}/{y}={result}")
    else: # This handles invalid input for choice
      print("Invalid Input!Please select a valid operation(1/2/3/4/5).")
calculator()
```

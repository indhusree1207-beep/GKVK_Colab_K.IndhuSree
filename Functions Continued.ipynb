```python
def average_yield(yields):
  total=sum(yields)
  return total/len(yields)
crop_yields=[1200,1350,1100,1500]
avg=average_yield(crop_yields)
print("Average Yield:",avg)
```

```python
def increase_yield(yields,increment):
    for i in range (len(yields)):
        yields[i]+=increment
crop_yields=[1200,1350,1100]
increase_yield(crop_yields,100)
print(crop_yields)
```

```python
def convert_to_quintals(yields):
  result=[]
  for value in yields:
    result.append(value/100)
  return result
crop_yield=[1200,1350,1500,1100]
quintals=convert_to_quintals(crop_yield)
print(quintals)
```

```python
def high_yield_crops(yields,threshold):
  high=[]
  for value in yields:
    if value>=threshold:
      high.append(value)
  return high
crop_yields=[1200,900,1500,800]
print(high_yield_crops(crop_yields,1000))
```

```python
def sort_prices(prices):
  prices.sort(reverse=True)
  return prices
crop_prices=[1800,1500,2200,2000]
print(sort_prices(crop_prices))
```

```python
def sort_prices(prices):
  prices.sort()
  crop_prices_sorted=sorted(prices)
  return crop_prices_sorted
crop_prices=[1800,1500,2200,2000]
new_prices=sort_prices(crop_prices)
print(new_prices)
```

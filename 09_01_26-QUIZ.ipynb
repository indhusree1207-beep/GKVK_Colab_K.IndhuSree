```python
costs=[30,45,25]
rates=[2,1.5,3]

recommendations=[]
for i in range(len(costs)):
  for j in range(len(rates)):
    if i!=j and costs[i]*rates[j]<80:
       recommendations.append(f"F{i+1}K{j+1}")
print(recommendations)
```

```python
lst=[1,2,-3,4]
def filter_positive():
  result=[]
  for num in lst:
    if num>0:
      result.append(num*2)
  return result
numbers=filter_positive()
print(numbers)
```

```python
my_list=[1,2,3]
def double_pairs(lst):
  for x in lst:
   if x%2==1:
    return[(x,x*2)]
our_list=double_pairs(my_list)
print(our_list)
```

```python
my_list=[1,2,3]
def double_pairs(lst):

  return[(x,x*2) for x in lst  if x%2==1]
our_list=double_pairs(my_list)
print(our_list)
```

```python
scans=[[1,0,1],[0,1,0],[1,0,0]]
alerts=[]
for i in range(len(scans)):
  row_count=0
  for j in range(len(scans[i])):
    if scans[i][j]==1:
      row_count+=1
    if row_count>1:
      alerts.append(f"Row{i} quarantine")
print(alerts)
```

```python
def analyze(lists):
  empty_count=0
  i=0
  while i<len(lists):
    if not lists[i]:  #if not lists means false
      empty_count+=1
    i+=1
    return empty_count
mylist=[[],[1],[2,3]]
ec=analyze(mylist)
print(ec)
```
